## Total Transactions
Filtered by : Genre and Artist

Tables:
- invoice line - invoice : InvoiceID
- invoice line - track : TrackID
- track - Genre : GenreID -> Genre
- Track - Album: AlbumID
- Album - Artist : ArtisID -> Name


```python
query = """
SELECT  
ivi.InvoiceLineId ,ivi.InvoiceID ,ivi.TrackId ,ivi.UnitPrice ,ivi.Quantity,
iv.InvoiceDate, iv.BillingCountry,
genres.Name as GenreName,
artists.Name as ArtistName

FROM invoice_items as ivi
LEFT JOIN invoices as iv ON ivi.InvoiceID = iv.InvoiceID
LEFT JOIN tracks as tr ON ivi.TrackID = tr.TrackID
LEFT JOIN genres ON tr.GenreId = genres.GenreId
LEFT JOIN albums ON tr.AlbumId = albums.AlbumId
LEFT JOIN artists ON albums.ArtistId = artists.ArtistId
"""
total_transactions_raw = pd.read_sql_query(query, connection)
total_transactions_raw['InvoiceDate'] = pd.to_datetime(total_transactions_raw['InvoiceDate'])
```

```python
filter_genre = genres['Name'].values.tolist() # all genres
filter_genre = ['Rock']
mask_genre = total_transactions_raw['GenreName'].apply(lambda x: x in filter_genre)

filter_artist = artists['Name'].values.tolist() # all artist
filter_artist = ['AC/DC']
mask_artist = total_transactions_raw['ArtistName'].apply(lambda x: x in filter_artist)

filter_year = range(2009,2010+1)
mask_year = total_transactions_raw['InvoiceDate'].apply(lambda x: x.year in filter_year)
```

```python
total_transactions_filtered = total_transactions_raw[(mask_genre) & (mask_artist) & (mask_year)]
total_transactions_filtered
```

manipulating datetime : https://www.dataquest.io/blog/python-datetime-tutorial/

[Read also `Pandas.DataFrame.resample` here](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.resample.html)

```python
invoice_date_range = pd.date_range(start=invoices['InvoiceDate'].min(),
                                   end=invoices['InvoiceDate'].max()
                                  )
```

```python
total_transactions = total_transactions_filtered.set_index('InvoiceDate')[['UnitPrice','Quantity']].resample('D').sum().reindex(invoice_date_range, fill_value=0).resample('Y', convention='end').sum()
```

# Aggregate and Pie


## Pie 

```python
def pie_top(df, column, threshold_count=5, threshold_percentage=0.05):
    df_count = df[column].value_counts()
    if len(df_count) <= threshold_count:
        return df_count
    else:
        df_prcnt = df_count/df_count.sum()
        mask = df_prcnt<threshold_percentage
        dff1 = df_count[~mask]
        dff2 = df_count[mask]
        if len(dff2) != 0 :
            other_count = dff2.sum()
            return dff1.append(pd.Series([other_count], index=['Other']))
        else:
            return dff1
    
```

```python
threshold_count = 5
threshold_percentage = 0.2
```
